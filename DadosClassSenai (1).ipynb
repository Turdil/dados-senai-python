```python
import pandas as pd
from google.colab import files
from IPython.display import display
import matplotlib.pyplot as plt

# Definindo as colunas
colunas = ["Nome", "Idade", "Preferência(Doce/Salgado)", "Cor Favorita", "Horas de Sono", "Gênero Musical"]

# Criar DataFrame vazio
tb = pd.DataFrame(columns=colunas)
display(tb)

# Exportar como CSV
arquivo_csv = "dados_modelo.csv"
tb.to_csv(arquivo_csv, index=False)

# Baixar o CSV vazio para preenchimento
files.download(arquivo_csv)
print("Baixe o arquivo CSV e preencha com os dados coletados!")
```

```python
# Upload do arquivo preenchido
subir = files.upload()

# Pegar o nome do arquivo enviado
nome_arquivo = list(subir.keys())[0]
print(f'Arquivo enviado: {nome_arquivo}')
```

```python
# Ler o CSV preenchido
dados = pd.read_csv(nome_arquivo)
display(dados)
```
