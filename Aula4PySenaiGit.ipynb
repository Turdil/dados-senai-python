```python
import pandas as pd
from IPython.display import display

import matplotlib.pyplot as plt
```

```python
dados = [
    {"nome": "João", "Idade": 30, "Destino": "São Paulo", "Voo": "AZ123"},
    {"nome": "Maria", "Idade": 25, "Destino": "Rio de Janeiro", "Voo": "BR456"},
    {"nome": "Pedro", "Idade": None, "Destino": "Belo Horizonte", "Voo": None},
    {"nome": "Marcos", "Idade":25, "Destino": "Ceara", "Voo": "LT789" },
    {"nome": "Vanessa", "Idade":26, "Destino": "Natal", "Voo": "XY321" },
    {"nome": "Marcos", "Idade":30, "Destino": "Recife", "Voo": "GH654" }

]
tb=pd.DataFrame(dados)
display(tb)
```

```python
tb_flitrada=tb.drop_duplicates()
display(tb_flitrada)
```

```python
tb["Idade"]=tb["Idade"].fillna(tb["Idade"].median())
tb["Voo"]=tb["Voo"].fillna("Não informado")
display(tb)

```

```python
tb["Idade"]=tb["Idade"].fillna(tb["Idade"].median())
display(tb)
```

```python
print(f'Média:{tb["Idade"].mean():.2f}')
```

```python
print(f'Destino:')
print(tb["Destino"].value_counts().plot(kind="bar", title="Destino"))
plt.ylabel("Quantidade de passageiros")
plt.xlabel("Destino")
plt.grid(True)
plt.show()
```
