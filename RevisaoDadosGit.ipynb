```python
import pandas as pd
from google.colab import files
from IPython.display import display
import matplotlib.pyplot as plt

colunas=["ID","Nome_do_Passageiro","Data_de_Embarque","Hora_de_Embarque","Número_do_Voo","Assento", "Destino"]

tb=pd.DataFrame(columns=colunas)
display(tb)

arquivo_excel="dados_modelos.xlsx"

tb.to_excel(arquivo_excel,index=False,engine="openpyxl")

files.download(arquivo_excel)
print("Baixe o arquivo e preencha corretament com os dados coletados")
```

```python
subir = files.upload()
nome_arquivo = list(subir.keys())[0]
print(f'Arquivo{nome_arquivo}')
```

```python
dados=pd.read_excel(nome_arquivo,engine="openpyxl")
display(dados)
```

```python
print("Cabeçalho")
display(dados.head())
```

```python
print("\n Numero de Voo")
print(dados["Número do Voo"].value_counts()) # Corrected column name
dados['Destino'].value_counts().plot(kind='bar',title="Destino")
plt.xlabel("Destino")
plt.ylabel("Numero de cidades")
plt.grid(True)
plt.show()
```
